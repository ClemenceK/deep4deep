# Embedding

```python
import numpy as np
```

```python
from gensim.models import Word2Vec
import csv 

my_embedding = Word2Vec(sentences=my_df['meta_description_preprocessed'], min_count=1, window=5, size=30, sample=1e-5)

# to use tensor flow 'auto-tSNE', need a TSV (tab-separated values) and a separate file of names

# building TSV for use in Embedding Projector

def embed_text(preprocessed_text):
    result = []
    for word in preprocessed_text:
            try:
                vector = my_embedding.wv.get_vector(word)
            except:
                vector = np.zeros(30)
            result.append(vector)
    return result

embed_text(my_df.meta_description_preprocessed[0])     
```

```python
my_df['embedded_meta_desc'] = my_df.meta_description_preprocessed.map(embed_text)
```

---
# tsv for t-SNE in embedding projector

```python
def flatten(nested_list):
    return [item for sublist in nested_list for item in sublist]
```

```python
with open('my_embedding.tsv', 'w') as vectors_tsvfile:
    with open('my_metadata.tsv', 'w') as metadata_tsvfile:
        writer1 = csv.writer(vectors_tsvfile, delimiter='\t')
        writer2 = csv.writer(metadata_tsvfile, delimiter='\t') 
        writer2.writerow(['company','industries','deeptech'])
        
        # calc minimum length of list
        tmp = [flatten(list(my_df.embedded_meta_desc[i])) for i in range(my_df.shape[0])]
        list_of_lengths = [len(tmp[i]) for i in range(len(tmp))]
        kept_len = min(list_of_lengths)
        
        for i in range(my_df.shape[0]):
            writer1.writerow(flatten(list(my_df.embedded_meta_desc[i]))[:kept_len])
            writer2.writerow([my_df.name[i], my_df.industries[i], my_df.target[i]])
```
